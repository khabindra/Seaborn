Barplot Python Seaborn

```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
%matplotlib inline 
```

```python
df = sns.load_dataset('penguins')
df.tail(100)
```

```python
sns.barplot(x=df.island,y=df.bill_length_mm)
plt.show()
```

```python
sns.barplot(x='island',y='bill_length_mm',data=df)
plt.show()
```

```python
sns.barplot(x='island',y='bill_length_mm',data=df,hue='sex')
plt.show()
```

```python
order_1 = ['Dream','Biscoe','Torgersen']
sns.barplot(x='island',y='bill_length_mm',data=df,hue='sex',order=order_1,
            hue_order=['Female','Male'])
plt.show()
```

```python
order_1 = ['Dream','Biscoe','Torgersen']
sns.barplot(x='island',y='bill_length_mm',data=df,hue='sex',order=order_1,
            hue_order=['Female','Male'],
            errorbar=('ci', 10) )
plt.show()
```

```python

sns.barplot(x='island',y='bill_length_mm',data=df,hue='sex',
            hue_order=['Female','Male'],
            errorbar=('ci', 10),n_boot= 4 )
plt.show()
```

```python
sns.barplot(x='bill_depth_mm',y='bill_length_mm',data=df,hue='sex',orient='h')
plt.show()
```

```python
sns.barplot(x='bill_depth_mm',y='bill_length_mm',data=df,hue='sex',orient='v')
plt.show()
```

```python
sns.barplot(x='island', y='bill_length_mm', data=df, hue='sex', palette=['red', 'blue'])
plt.show()
```

```python
sns.barplot(x='island', y='bill_length_mm', data=df, color='red')
plt.show()
```

```python
sns.barplot(x='island', y='bill_length_mm', data=df, hue='sex', palette='icefire',saturation=50)
plt.show()
```

```python
sns.barplot(x='island', y='bill_length_mm', data=df, hue='sex',
             err_kws={'color': 'b', 'linewidth': 10})
plt.show()
```

```python
sns.barplot(x='island', y='bill_length_mm', data=df, hue='sex',
            hue_order=['Female', 'Male'], saturation=100, capsize=0.5,
            errorbar=('ci', 10), n_boot=4, err_kws={'color': 'b'},alpha=0.9)
plt.show()
```

```python
sns.set(style='dark')
sns.barplot(x='island', y='bill_length_mm', data=df, hue='sex',
            hue_order=['Female', 'Male'], saturation=100, capsize=0.5,
            errorbar=('ci', 10), n_boot=4, err_kws={'color': 'b'},alpha=0.9)
plt.show()
```
