Seaborn Scatter Plot

```python
import seaborn as sns 
import matplotlib.pyplot as plt 
import pandas as pd 
```

```python
df = sns.load_dataset('penguins')
df.head()
```

```python
sns.scatterplot(x='bill_length_mm',y='bill_depth_mm',data=df)
plt.show()
```

```python
sns.scatterplot(x='bill_length_mm',y='bill_depth_mm',data=df,color='g')
plt.show()
```

```python
sns.scatterplot(x='bill_length_mm',y='bill_depth_mm',data=df,hue='sex',palette=['red','blue'])
plt.show()
```

```python
sns.scatterplot(x='bill_length_mm',y='bill_depth_mm',data=df,hue='sex',palette='Accent')
plt.show()
```

```python
sns.scatterplot(x='bill_length_mm',y='bill_depth_mm',data=df,hue='sex',palette=['red','blue'],style='sex')
plt.show()
```

```python
sns.scatterplot(x='bill_length_mm',y='bill_depth_mm',data=df,hue='sex',palette='Accent',style='sex')
plt.show()
```

```python
sns.scatterplot(x='bill_length_mm',y='bill_depth_mm',data=df.head(100),hue='sex',palette=['red','blue'],style='sex',size='sex',sizes=(80,30))
plt.show()
```

```python
sns.scatterplot(x='bill_length_mm',y='bill_depth_mm',data=df.head(100),hue='sex',palette='gist_ncar',style='sex',size='sex',sizes=(80,30))
plt.show()
```

```python
sns.scatterplot(x='bill_length_mm',y='bill_depth_mm',data=df.head(100),hue='sex',palette='gist_ncar',style='sex',size='sex',sizes=(80,30),alpha=0.1)
plt.show()
```

```python
sns.scatterplot(x='bill_length_mm',y='bill_depth_mm',data=df.head(100),hue='sex',palette='gist_ncar',style='sex',size='sex',sizes=(80,30),alpha=0.5)
plt.show()
```

```python
sns.scatterplot(x='bill_length_mm',y='bill_depth_mm',data=df.head(100),hue='sex',palette='gist_ncar',style='sex',size='sex',sizes=(80,30),alpha=1)
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.style.use('dark_background') 
plt.grid()
m = {'Male':'*','Female':'o'}
sns.scatterplot(x='bill_length_mm',y='bill_depth_mm',data=df.head(150),hue='sex',palette='gist_ncar',style='sex',size='sex',sizes=(50,80),alpha=1,markers=m)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
plt.style.use('dark_background')  # Set the background style to dark

# Define markers for different sexes
markers = {'Male': '*', 'Female': 'o'}

# Create the scatter plot
sns.scatterplot(x='bill_length_mm', y='bill_depth_mm', data=df.head(150), hue='sex', palette='gist_ncar', style='sex', size='sex', sizes=(50, 80), alpha=1, markers=markers)

plt.grid(color='white', linestyle='--', linewidth=0.5)  # Set the grid color to white
plt.show()
```
