Histogram Plot 

```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
df = sns.load_dataset('penguins')
df
```

```python
sns.displot(df['bill_length_mm'])
plt.show()
```

```python
sns.displot(df['flipper_length_mm'])
plt.show()
```

```python
sns.displot(df['flipper_length_mm'],bins=[170,180,190,200,210,220,230,240])
plt.show()
```

```python
sns.displot(df['flipper_length_mm'],bins=[170,180,190,200,210,220,230,240],kde=True)
plt.show()
```

```python
sns.displot(df['flipper_length_mm'],bins=[170,180,190,200,210,220,230,240],kde=True,rug=True)
plt.show()
```

```python
sns.displot(df['flipper_length_mm'],bins=[170,180,190,200,210,220,230,240],kde=True,rug=True,color='g')
plt.show()
```
