HeatMap 

```python
import matplotlib.pyplot as plt 
import pandas as pd 
import seaborn as sns 
import numpy as np 
```

```python
var = np.linspace(1,10,20).reshape(4,5)
var
```

```python
sns.heatmap(var)
plt.show()
```

```python
sns.heatmap(var,vmin=0,vmax=8)
plt.show()
```

```python
sns.heatmap(var,vmin=0,vmax=8,cmap='PuOr')
plt.show()
```

```python
sns.heatmap(var,vmin=0,vmax=8,cmap='PuOr',annot=True)
plt.show()
```

```python
var1 = np.linspace(1,10,10).reshape(2,5)
var1
```

```python
sns.heatmap(var1,vmin=0,vmax=10,cmap='PuOr',annot=True)
plt.show()
```

```python
ar = np.array([['a0','a1','a2','a3','a4'],
               ['b0','b1','b2','b3','b4']])
ar
```

```python
y = {'fontsize':20,'color':'r'}
sns.heatmap(var1,vmin=0,vmax=10,cmap='PuOr',annot=ar,fmt='s',annot_kws=y)
```

```python
y = {'fontsize':20,'color':'r'}
sns.heatmap(var1,vmin=0,vmax=10,cmap='PuOr',annot=ar,fmt='s',annot_kws=y,linewidth=10)
```

```python
y = {'fontsize':20,'color':'r'}
sns.heatmap(var1,vmin=0,vmax=10,cmap='PuOr',annot=ar,fmt='s',annot_kws=y,linewidth=10,linecolor='y')
```

```python
y = {'fontsize':20,'color':'r'}
sns.heatmap(var1,vmin=0,vmax=10,cmap='PuOr',annot=ar,fmt='s',annot_kws=y,linewidth=10,linecolor='y',cbar=False)
plt.show()
```

```python
y = {'fontsize':20,'color':'r'}
sns.heatmap(var1,vmin=0,vmax=10,cmap='PuOr',annot=ar,fmt='s',annot_kws=y,linewidth=10,linecolor='y',cbar=False,xticklabels=False)
plt.show()
```

```python
y = {'fontsize':20,'color':'r'}
sns.heatmap(var1,vmin=0,vmax=10,cmap='PuOr',annot=ar,fmt='s',annot_kws=y,linewidth=10,linecolor='y',cbar=False,yticklabels=False)
plt.show()
```

```python
y = {'fontsize': 12, 'color': 'r'}
v = sns.heatmap(var1, vmin=0, vmax=10, cmap='PuOr', annot=ar, fmt='s', annot_kws=y, linewidth=10, linecolor='y', cbar=False, yticklabels=False)
v.set(xlabel='python', ylabel='education')
# Set the font scale to 1.2 (you can adjust this value as needed)
sns.set(font_scale=1.2)
plt.show()
```
