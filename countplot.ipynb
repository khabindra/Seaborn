Count Plot 

```python
import matplotlib.pyplot as plt 
import seaborn as sns 
import pandas as pd 
```

```python
var = sns.load_dataset('tips')
var
```

```python
sns.countplot(x='sex',data=var,color='r')
plt.show()
```

```python
sns.barplot(x='sex',y='size',data=var)
plt.show()
```

```python
sns.countplot(x='sex', data=var, palette='dark:r', hue='smoker')
plt.show()
```

```python
sns.countplot(y='sex', data=var, palette='dark:r', hue='smoker')
plt.show()
```

```python
sns.countplot(x='sex', data=var, palette='bwr', hue='smoker')
plt.show()
```

```python
sns.countplot(x='sex', data=var, hue='smoker', palette='dark:y')
plt.show()
```

```python
sns.countplot(x='sex', data=var, hue='smoker', palette='dark:y',saturation=0.1)
plt.show()
```

```python
sns.countplot(x='sex', data=var, hue='smoker', palette='dark:y',saturation=0.5)
plt.show()
```

```python
sns.countplot(x='sex', data=var, hue='smoker', palette='dark:y',saturation=1)
plt.show()
```
