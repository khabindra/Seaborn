Line Plot

```python
import seaborn as sns 
import matplotlib.pyplot as plt 
import pandas as pd 
```

```python
var = [1,2,3,4,5,6,7]
var_1 = [2,3,4,1,5,6,7]
x_1 = pd.DataFrame({'var':var,
                  'var_2':var_1})

print(x_1)
print('\nline plot using seaborn: ')
sns.lineplot(data=x_1, x='var', y='var_2')
# Show the plot
plt.show()
```

```python
print('Line plot using matplotlib')
plt.plot(var,var_1)
plt.show()
```

```python
print('line plot using seaborn in the form of matplotlib :')
sns.lineplot(x=var,y=var_1)
plt.show()
```

```python
y_1 = sns.load_dataset('penguins')
y_1
```

```python
print('\nline plot using seaborn: ')
sns.lineplot(data=y_1, x='bill_length_mm', y='flipper_length_mm')
# Show the plot
plt.show()
```

```python
print('\nline plot using seaborn: ')
sns.lineplot(data=y_1, x='bill_length_mm', y='flipper_length_mm',hue='sex')
# Show the plot
plt.show()
```

```python
print('\nline plot using seaborn: ')
sns.lineplot(data=y_1, x='bill_length_mm', y='flipper_length_mm',hue='sex',size=10)
# Show the plot
plt.show()
```

```python
print('\nline plot using seaborn: ')
sns.lineplot(data=y_1, x='bill_length_mm', y='flipper_length_mm',hue='sex',size=10,style='sex')
# Show the plot
plt.show()
```

```python
print('\nline plot using seaborn: ')
sns.lineplot(data=y_1, x='bill_length_mm', y='bill_depth_mm',hue='sex',size=10,style='sex')
# Show the plot
plt.show()
```

```python
print('\nline plot using seaborn: ')
sns.lineplot(data=y_1, x='bill_length_mm', y='bill_depth_mm',hue='sex',size=10,style='sex',palette='Accent')
# Show the plot
plt.show()
```

```python
print('\nline plot using seaborn: ')
sns.lineplot(data=y_1, x='bill_length_mm', y='bill_depth_mm',hue='sex',size=10,style='sex',palette='Accent',markers=['o','>'])
# Show the plot
plt.show()
```

```python
y_1 = y_1.head(100)
y_1
```

```python
print('\nline plot using seaborn: ')
sns.lineplot(data=y_1, x='bill_length_mm', y='bill_depth_mm',hue='sex',
             size=10,style='sex',palette='Accent',markers=['o','>'],
             dashes=False)
# Show the plot
plt.show()
```

```python
print('\nline plot using seaborn: ')
sns.lineplot(data=y_1, x='bill_length_mm', y='bill_depth_mm',hue='sex',
             size=10,style='sex',palette='Accent',markers=['o','>'],
             dashes=False,legend=False)
# Show the plot
plt.show()
```

```python
print('\nline plot using seaborn: ')
sns.lineplot(data=y_1, x='bill_length_mm', y='bill_depth_mm',hue='sex',
             size=10,style='sex',palette='Accent',markers=['o','>'],
             dashes=False,legend='full')
# Show the plot
plt.show()
```

```python
print('\nline plot using seaborn: ')
sns.lineplot(data=y_1, x='bill_length_mm', y='bill_depth_mm',hue='sex',
             size=10,style='sex',palette='Accent',markers=['o','>'],
             dashes=False,legend='full')
plt.grid()
plt.title('Python',fontsize=20)
# Show the plot
plt.show()
```
